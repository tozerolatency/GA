```python
neg_corr_df = neg_corr_df.loc['2000-01-01':'2009-01-01']
##neg_corr_df = neg_corr_df.loc['2009-01-01':'2009-01-01']
#neg_corr_df = neg_corr_df.loc['2009-01-01':'2018-07-01']
neg_corr_df.head()
```

```python
neg_corr_returns = (neg_corr_df / neg_corr_df.shift(1)) - 1
neg_corr_returns.head()
```

# Machine Learning

#### get the stock data from yahoo finance

```python
#df = wb.DataReader("PG", data_source='yahoo', start='2000-1-1')  # sharpe 1.59
df = wb.DataReader("IBM", data_source='yahoo', start='2000-1-1')  # sharpe 1.76
```

#### Query to yahoo finance api fails sometimes, just keep trying couple of times if it fails

```python
df.head()
```

```python
df['Return']= np.log(df['Close']/df['Close'].shift(1))
df.dropna(how='any',axis=0,inplace=True)
df.head()
```

```python
ohlc = df[['Open','High','Low','Close']]
ohlc.head()
```

```python
ohlc['open'] =ohlc['Open'].shift(1)
ohlc['high'] =ohlc['High'].shift(1)
ohlc['low']  =ohlc['Low'].shift(1)
ohlc['close']=ohlc['Close'].shift(1)

ohlc.head()

```

```python
ohlc=ohlc[['open','high','low','close']]
ohlc.head()
```

```python
ohlc=ohlc.dropna()

ohlc.head()
```

### Using unsupervised learning algorithm to generate feature called category

```python
model = mix.GaussianMixture(n_components=4, covariance_type="spherical", n_init=200, random_state=42)
model.fit(np.reshape(ohlc,(-1,ohlc.shape[1])))
```

```python
category_pred = model.predict(np.reshape(ohlc,(-1,ohlc.shape[1])))
```

```python
ohlc['Return']= np.log(ohlc['close']/ohlc['close'].shift(1))
ohlc.head()
```

#### Create new columns in your dataframe
#### column 'category' is created from GaussianMix algorithm

```python
categories=pd.DataFrame(category_pred,columns=['category'],index=ohlc.index).join(ohlc, how='inner')
categories['market_cumm_return']=ohlc.Return.cumsum()
categories.reset_index(drop=False,inplace=True)
categories.rename(columns={'index':'Date'})
categories.head()
```

```python
categories.category.unique()
```

```python
category_list=[0,1,2,3]
fig = sns.FacetGrid(data=categories,hue='category',hue_order=category_list,aspect=2,size= 4)
fig.map(plt.scatter,'Date','market_cumm_return', s=4).add_legend()
plt.show()
```

#### check the mean and covariance for categories

```python
for i in category_list:
    print('Mean for regime %i: '%i,model.means_[i][0])
    print('Co-Variancefor regime %i: '%i,(model.covariances_[i]))
```

```python
categories.head()
```

```python
categories.dropna(inplace=True,how='any',axis=0)
categories.head()
```

```python
categories['Signal'] = 0
categories.loc[categories['Return']>0,'Signal']=1
categories.loc[categories['Return']<0,'Signal']=-1

categories.head()
```

### using StandardScaler for Standardization

```python
ss =StandardScaler()
columns =categories.columns.drop(['category','Date','Return','market_cumm_return','Signal'])    
columns
```

```python
categories[columns]= ss.fit_transform(categories[columns])
categories.head()
```

### using SVM (support vector machines) classifier to predict the signal/output variable.

```python
svm_classifier = SVC(C=1.0,decision_function_shape=None, degree=3, gamma='auto', kernel='rbf')
```

#### split the data into 70% training and 30% test data

```python
split_size = 0.7

row_num = int( split_size * len(categories))

print (len(categories))
print (row_num)
```

```python
X = categories.drop(['Signal','Return','market_cumm_return','Date'], axis=1)
y= categories['Signal']
```

#### fit the model

```python
X_train=X[:row_num]
y_train=y[:row_num]

svm_classifier.fit(X_train,y_train)
```

#### model evaluation

```python
df['strategy_return'] = df['Signal_Predicted']*df['Return'].shift(-1)
df['strategy_cumm_return']=0
df['market_cumm_return']=0
```

```python
df.iloc[-test_row_num:,df.columns.get_loc('strategy_cumm_return')] \
       = np.nancumsum(df['strategy_return'][-test_row_num:])
    
df.iloc[-test_row_num:,df.columns.get_loc('market_cumm_return')] \
       = np.nancumsum(df['Return'][-test_row_num:])
    
Sharpe = (df['strategy_cumm_return'][-1]-df['market_cumm_return'][-1])\
           /np.nanstd(df['strategy_cumm_return'][-test_row_num:])

```

```python
plt.figure(figsize=(15,10))
plt.plot(df['strategy_cumm_return'][-test_row_num:],color='g',label='Strategy Returns')
plt.plot(df['market_cumm_return'][-test_row_num:],color='r',label='Market Returns')
plt.figtext(0.14,0.9,s='Sharpe ratio: %.2f'%Sharpe)
plt.legend(loc='best')
plt.show()
```

```python
#output the hit-rate and the confusion matrix for the model
print("Accuracy of SVM model: %0.3f" % svm_classifier.score(X_test, y_test))
print("Confusion matrix: \n%s" % confusion_matrix(df.iloc[-test_row_num:,df.columns.get_loc('Signal_Predicted')], y_test))
```
